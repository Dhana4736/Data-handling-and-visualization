```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[4,5,5]
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
import numpy as np
x=np.array([10,20,30,40])
mylabels=["apples","bananas","cherries","dates"]
mycolors=["black","hotpink","b","green"]
explode=[0,0.1,0.1,0.1]
plt.pie(x,labels=mylabels,colors=mycolors,explode=explode,shadow="true")

plt.show()
```

```python
x=[10,20,30,40]
plt.subplot(2,2,4)
plt.pie(x)
plt.show(x)
```

```python
x=[1,2,3,4]
y=[1,2,3,4]
color=["red","green","blue","orange"]
sizes=[100,200,300,400]
plt.title("Scatter graph")
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.scatter(x,y,color=color,marker="*",alpha=1,s=sizes,edgecolor="g")
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[1,2,3,4]
plt.subplot(2,2,1)
plt.plot(x,y)

x=[10,20,30,40]
plt.subplot(2,2,2)
plt.pie(x)
plt.show(x)

x=[1,2,3,4]
y=[1,2,3,4]
plt.subplot(2,2,3)
plt.bar(x,y)

x=[1,2,3,4]
y=[1,2,3,4]
plt.subplot(2,2,4)
color=["red","green","blue","orange"]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar graph")
plt.bar(x,y,color=color)
```
