```python
import folium

locations = {
    "MG Road": [12.9716, 77.5946],
    "Brigade Road": [12.9796, 77.6034],
    "Church Street": [12.9766, 77.5906],
    "Lalbagh": [12.9516, 77.5866],
    "Cubbon Park": [12.9736, 77.5986],
}

mymap = folium.Map(location=[12.9716, 77.5946],
                    width=950,
                    zoom_start=12,
                    tiles='openstreetmap')

for name, coordinates in locations.items():
    marker = folium.Marker(location=coordinates, popup=folium.Popup(name))
    mymap.add_child(marker)

mymap
```

```python
import folium

locations = {

    "Vidana soudha": {
        "coordinates": [12.9796, 77.5908],
        "image":"https://bengaluruprayana.com/wp-content/uploads/2023/09/Vidhana-Soudha-Bengaluru.jpeg"
    },
    "Church Street": {
        "coordinates": [12.9766, 77.5906],
        "image": "https://i.ytimg.com/vi/ft3ihb6ie5M/sddefault.jpg"
    },
    "Lalbagh": {
        "coordinates": [12.9516, 77.5866],
        "image": "https://www.theleela.com/prod/content/assets/styles/tl_1920_735/public/aio-banner/dekstop/Lalbagh-Bangalore_2.jpg?VersionId=QSNt6JcPWLsk5GCcTqUHKvu6nlMht7mZ&itok=AsR-vSyP"
    },
    "Cubbon Park": {
        "coordinates": [12.9736, 77.5986],
        "image": "https://bengaluruprayana.com/wp-content/uploads/2023/02/Cubbon-Park.jpeg"
    },
}

mymap = folium.Map(location=[12.9716, 77.5946],
                    width=950,
                    zoom_start=12,
                    tiles='openstreetmap')

for name, location_info in locations.items():
    coordinates = location_info["coordinates"]
    image_url = location_info["image"]
    marker = folium.Marker(location=coordinates, popup=folium.Popup(html=f'<img src="{image_url}">'))
    mymap.add_child(marker)

mymap
```
